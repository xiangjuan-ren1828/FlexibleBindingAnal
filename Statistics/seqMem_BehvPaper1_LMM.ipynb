```r
emm1 = emmeans(choiceModel, specs = ~ trialNo*measNo*groupNo) # same as: emmeans(choiceModel, ~ networkName) & emmeans(choiceModel, c("networkName"))
emm1
```

```r
emm_pairs <- pairs(emm1, simple = "measNo", adjust = "bonferroni") # adjust = "bonferroni", "tukey"
emm_pairs
```

### Memory gradient: partial retrieval from three-retrieval trials and full retrieval from the first reporting window

```r
# ------ Three reports: item and location within trials, reconstruction from the 1st report window (between trials) ------
# Subset for partial reports
acc_glmm_pd_partialReports <- acc_glmm_pd[
  acc_glmm_pd$dataType == "0withinTrl" &
  (acc_glmm_pd$reportType == "0content" | acc_glmm_pd$reportType == "1position"),
]
# Subset for full reports (between trials)
acc_glmm_pd_fullReportsBetween <- acc_glmm_pd[
  acc_glmm_pd$dataType == "1betweenTrl" &
  acc_glmm_pd$reportType == "2both",
]
# Combine them
acc_glmm_pd_combined <- rbind(
  acc_glmm_pd_partialReports,
  acc_glmm_pd_fullReportsBetween
)

acc_glmm_pd_combined$trialNo = scale(acc_glmm_pd_combined$trialNo, center=TRUE, scale=FALSE) #scale=FALSE
acc_glmm_pd_combined
```

```r
choiceModel <- mixed(resp ~ reportType*ageNo + (1 | subjNo), data = acc_glmm_pd_combined, family = binomial, method="LRT")
choiceModel
```

```r
emm1 = emmeans(choiceModel, specs = ~ reportType*ageNo)
emm1
```

```r
emm_pairs <- pairs(emm1, simple = "reportType", adjust = "bonferroni") # adjust = "bonferroni", "tukey"
emm_pairs
```
