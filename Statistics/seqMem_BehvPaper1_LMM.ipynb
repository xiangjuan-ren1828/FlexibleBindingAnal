## Figure 2C: accuracy comparison across 3 measures (itemReport, posReport, reconsReport) between across-window and within-window analysis in Experiment 2 

```r
path_or_buf  <- paste0(statSaveDir, 'acc_fullyCor_within_between_trials_YA_OA_group_exp2.csv' )
acc_glmm_pd <- read.csv(file = path_or_buf)
acc_glmm_pd
```

```r
# ------ Three reports within a trial ------
acc_glmm_pd_threeReports = acc_glmm_pd[acc_glmm_pd$groupNo == '0interleaved' & acc_glmm_pd$dataType == '0withinTrl', ]
acc_glmm_pd_threeReports
```

```r
# ------ Three reports: item and location within trials, reconstruction from the 1st report window (between trials) ------
# Subset for partial reports
acc_glmm_pd_partialReports <- acc_glmm_pd[
  acc_glmm_pd$groupNo == "0interleaved" &
  acc_glmm_pd$dataType == "0withinTrl" &
  (acc_glmm_pd$reportType == "0content" | acc_glmm_pd$reportType == "1position"),
]

# Subset for full reports (between trials)
acc_glmm_pd_fullReportsBetween <- acc_glmm_pd[
  acc_glmm_pd$groupNo == "0interleaved" &
  acc_glmm_pd$dataType == "1betweenTrl" &
  acc_glmm_pd$reportType == "2both",
]

# Combine them
acc_glmm_pd_threeReports <- rbind(
  acc_glmm_pd_partialReports,
  acc_glmm_pd_fullReportsBetween
)
```

```r
acc_glmm_pd_threeReports
```

```r
choiceModel <- mixed(fullyCor ~ ageNo*reportType + (1 | subjNo), data = acc_glmm_pd_threeReports, family = binomial, method="LRT")
choiceModel
```

```r
emm1 = emmeans(choiceModel, specs = ~ ageNo*reportType)
emm1
```

```r
emm_pairs <- pairs(emm1, simple = "reportType", adjust = "bonferroni") # adjust = "bonferroni", "tukey"
emm_pairs
```
